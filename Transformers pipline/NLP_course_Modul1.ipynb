Тематическая классификация

```python
cls_pl = pipeline("text-classification", "nateraw/bert-base-uncased-ag-news")
cls_pl('NLP is a field of linguistics and machine learning focused on understanding everything related to human language.')
```

Token classification

```python
ner_pipeline = pipeline("ner", "dslim/bert-base-NER", aggregation_strategy="first")
ner_pipeline("Matthew Carrigan is a Machine Learning Engineer at Hugging Face. He lives in Dublin, Ireland and previously worked as an ML engineer at Parse.ly and before that as a post-doctoral researcher at Trinity College Dublin.")
```

Перевод с русского на английский

Аннотирование

```python
sum_pipeline = pipeline("summarization", "IlyaGusev/rut5_base_sum_gazeta")
sum_pipeline("Естественным развитием задач классификации всего текста или его фрагментов является классификация отдельных слов в предложении. Наиболее распространенной задачей такого вида является задача выделения именованных сущностей (NER - named entity recognition) - в данном случае необходимо для каждого слова в тексте проставить класс принадлежности его к тому или иному типу сущности (например, организация, персона, локация и т.д.), либо пустой класс, показывающий, что слово не относится ни к одному из рассматриваемых классов. С помощью NER можно разрабатывать системы автоматического заполнения форм документов, составления отчетов о компаниях, нахождения разных упоминания одних и тех же сущностей (это задача entity linking - она решается после ner), удаления персональной информации из текстов и т.д. Кроме классификации слов текста на именованные сущности, можно производить классификацию по каким-то другим критериям - например, по частям речи (хотя это практическое использование такого рода задач гораздо уже).")
```

Поиск ответа на вопрос

```python
qa_pipeline = pipeline("question-answering", "deepset/roberta-base-squad2")
question = "Which name is also used to describe the Amazon rainforest in English?"
context = "The Amazon rainforest (Portuguese: Floresta Amazônica or Amazônia; Spanish: Selva Amazónica, Amazonía or usually Amazonia; French: Forêt amazonienne; Dutch: Amazoneregenwoud), also known in English as Amazonia or the Amazon Jungle, is a moist broadleaf forest that covers most of the Amazon basin of South America. This basin encompasses 7,000,000 square kilometres (2,700,000 sq mi), of which 5,500,000 square kilometres (2,100,000 sq mi) are covered by the rainforest. This region includes territory belonging to nine nations. The majority of the forest is contained within Brazil, with 60% of the rainforest, followed by Peru with 13%, Colombia with 10%, and with minor amounts in Venezuela, Ecuador, Bolivia, Guyana, Suriname and French Guiana. States or departments in four nations contain \"Amazonas\" in their names. The Amazon represents over half of the planet's remaining rainforests, and comprises the largest and most biodiverse tract of tropical rainforest in the world, with an estimated 390 billion individual trees divided into 16,000 species."
qa_pipeline(question, context)
```

```python
qa_pl = pipeline("question-answering", "AlexKay/xlm-roberta-large-qa-multilingual-finedtuned-ru")
```

```python
question_1 = "Какая задача может быть развитием задачи классификации текстов?"
context_1 = "Естественным развитием задач классификации всего текста или его фрагментов является классификация отдельных слов в предложении. Наиболее распространенная задача такого вида — задача выделения именованных сущностей (NER - named entity recognition) — в этом случае необходимо для каждого слова в тексте проставить класс принадлежности его к тому или иному типу сущности (например, организация, персона, локация и т.д.), либо пустой класс, показывающий, что слово не относится ни к одному из рассматриваемых классов."
qa_pl(question_1, context_1)
```

```python
question_2 = "Что послужило основным толчком для существенного улучшения качества решения NLP задач в последние несколько лет?"
context_2 ="Для применения какого-либо метода машинного обучения с учителем (gradient boosting, support vector machines, etc.) необходимо сформировать признаковое пространство, в котором представлены объекты обучающей и тестовой выборки. В случае табличных данных мы в явном виде имеем матрицу объект-признак. Однако, необходимо привести в числовой вид переменные, значения которых в данных представлено не числами — вам известны несколько способов кодирования такого рода переменных (one-hot encoding, label encoding, embeddings, etc.). Рассмотрим теперь подробнее задачу классификации текстов — глобально она не сильно отличается от классификации табличных данных. Нам также необходимо сформировать признаковое пространство, а затем обучить некоторый алгоритм классификации. Здесь наша основная задача — понять, как можно представить текст в виде вектора, чтобы можно было обучить алгоритм машинного обучения. При этом важно не просто представить текст в виде вектора, а сделать это так, чтобы учитывались особенности естественных языков, в частности, многозначность слов, наличие подразумеваемых слов и т.д.Именно эта задача является одной из основных в NLP, и прорыв в качестве решения NLP задач в последние 3-4 года обусловлен именно новыми подходами к отображению текстов в векторное пространство."
qa_pl(question_2, context_2)
```

```python
question_3 = "Что служит входными данными для задачи поиска ответа на вопрос в заданном контексте?"
context_3 = "Еще одной из NLP задач, которой занимаются уже не одно десятилетие, является задача поиска ответов на сформулированный пользователем вопрос. Эта задача может быть представлена в некоторых вариациях, например, поиск ответа может осуществляться в какой-то базе знаний, либо в передаваемом пользователем контексте. Рассмотрим подробнее задачу в постановке поиска ответа на вопрос в заданном контексте. Если для рассмотренных ранее задач входными данными были только тексты, которые необходимо проанализировать, то в данном случае на вход подается два текста — вопрос и контекст, в котором необходимо найти ответ на вопрос (или, в некоторых случаях, указать, что его нет). Задача поиска ответов на вопросы используется, например, в голосовых помощниках. Приведем пример инференса для данной задачи с использованием уже рассмотренного класса pipeline. Для поиска нужных моделей нам понадобятся модели, обученные на соответсвующих датасетах, например, SQuAD. В качестве значения аргумента task используем “question-answering”, а в качестве модели выберем, например, “deepset/roberta-base-squad2”."
qa_pl(question_3, context_3)
```
